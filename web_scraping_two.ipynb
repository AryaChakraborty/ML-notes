### link = https://www.ambitionbox.com/list-of-companies

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36'}
data = pd.DataFrame()

for i in range(1,30) :
    web_page = requests.get('https://www.ambitionbox.com/list-of-companies?page={}'.format(i), headers=headers).text
    soup = BeautifulSoup(web_page, 'lxml')
    companies = soup.find_all('div', class_ = 'company-content-wrapper')
    
    name = []
    rating = []
    reviews = []
    ctype = []
    hq = []
    age = []
    employees = []
    about = []

    for company in companies : 
        if len(company.find_all('p', class_='infoEntity')) > 3 :
            name.append(company.find('h2').text.strip())
            rating.append(company.find('p', class_='rating').text.strip())
            reviews.append(company.find('a', class_='review-count').text.strip())
            ctype.append(company.find_all('p', class_='infoEntity')[0].text.strip())
            hq.append(company.find_all('p', class_='infoEntity')[1].text.strip())
            age.append(company.find_all('p', class_='infoEntity')[2].text.strip())
            employees.append(company.find_all('p', class_='infoEntity')[3].text.strip())
            about.append(company.find('p', class_='description').text.strip())
    
    diction = {'name' : name, 'rating' : rating, 'reviews' : reviews, 'ctype' : ctype, 'hq' : hq, 'age' : age, 'employees' : employees, 'about' : about}
    curr_data = pd.DataFrame(diction)
    data = data.append(curr_data, ignore_index = True)
```
