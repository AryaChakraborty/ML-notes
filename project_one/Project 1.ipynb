```python
sl = df.sl
sl = np.array(sl)
sw = df.sw
sw = np.array(sw)
pl = df.pl
pl = np.array(pl)
pw = df.pw
pw = np.array(pw)
```

```python
def entropy_feat(arr, y_train) : 
    dict = {}
    summm = 0
    for x in range(0, len(arr), 1) :
        i = arr[x]
        j = y_train[x]
        if(i in dict.keys()) :
            if(j in dict[i].keys()) :
                dict[i][j] = dict[i][j]+1
            else :
                dict[i][j] = 1
        else :
            dict[i] = {}
            dict[i][j] = 1
    
    list = []
    for x in dict.keys() :
        val = dict[x].values()
        s = sum(val) # sum of all cardinality
        en = 0
        for k in val :
            en = en + (k/s)*math.log((k/s), 2)
        list.append([s, en])
        summm = summm + s

    grand = 0
    for i in list :
        grand = grand + (i[0]/summm)*i[1]
    return grand # summation
```

```python
def split_index(arr, y_train) : 
    dict = {}
    summm = 0
    for x in range(0, len(arr), 1) :
        i = arr[x]
        j = y_train[x]
        if(i in dict.keys()) :
            if(j in dict[i].keys()) :
                dict[i][j] = dict[i][j]+1
            else :
                dict[i][j] = 1
        else :
            dict[i] = {}
            dict[i][j] = 1
    
    sum_list = [] # list of sum of all cardinality of a value in feature
    for x in dict.keys() :
        val = dict[x].values()
        s = sum(val) # sum of all cardinality
        sum_list.append(s)
        summm = summm + s
    
    split = 0
    for it in sum_list :
        split = split + (it/summm)*math.log((it/summm), 2)
    return -1*split
```

```python
diction = {}
for i in y_train :
    if(i in diction.keys()) :
        diction[i]=diction[i]+1
    else :
        diction[i]=1
val = diction.values()
summ = sum(val)
entropy = 0
for i in val :
    entropy = entropy + (i/summ)*math.log((i/summ), 2)
    
fin_entropy = entropy
```

```python
# x_train, y_train, feat
# f

sl_entropy = entropy_feat(sl, y_train)
sw_entropy = entropy_feat(sw, y_train)
pl_entropy = entropy_feat(pl, y_train)
pw_entropy = entropy_feat(pw, y_train)

sl_gain = abs(fin_entropy-sl_entropy)
sw_gain = abs(fin_entropy-sw_entropy)
pl_gain = abs(fin_entropy-pl_entropy)
pw_gain = abs(fin_entropy-pw_entropy)

max_ele = max(sl_gain, sw_gain, pl_gain, pw_gain)
print(sl_gain, sw_gain, pl_gain, pw_gain)
dec_feat = 'pl'
print(max_ele, dec_feat)
```

```python
sl_split = split_index(sl, y_train)
sw_split = split_index(sw, y_train)
pl_split = split_index(pl, y_train)
pw_split = split_index(pw, y_train)
# print(sl_split, sw_split, pl_split, pw_split)

sl_gain_r = sl_gain/sl_split
sw_gain_r = sw_gain/sw_split
pl_gain_r = pl_gain/pl_split
pw_gain_r = pw_gain/pw_split

feat = [[sl_gain_r, "sl"], [sw_gain_r, "sw"], [pl_gain_r, "pl"], [pw_gain_r, "pw"]]
# sort
feat.sort(reverse = True)
print(feat)
```

```python
def ispure(arr) :
    dict = {}
    for it in arr :
        if(it in dict.keys()) :
            dict[i] = dict[i] + 1
        else :
            dict[i] = 1
    
    for it in dict.keys() :
        if(dict[i] == len(arr)) :
            return True    
    return False
```

```python
# function : either leaf...'entropy', 'count', level = 0...while calling, we'll increment level
# decision(y_train, 0)
def decision(arr, idx) :
    # leaf node check
    array1 = [row[0] for row in arr] # y values
    array2 = [row[1] for row in arr] # index values
    
    if(ispure(array1) or idx==len(feat)) :
        return;
    # divide
    curr_feat_n = feat[idx][1]
    curr_feat = globals()[curr_feat_n]
    dit = {}
    for i in array2 :
        x = curr_feat[i]
        y = y_train[i]
        if(x in dit.keys()) :
            dit[x].append([y, i])
        else :
            dit[x] = [[y, i]]
    print("level = ", idx-1)
    s1 = curr_feat_n + "_entropy"
    p1 = globals()[s1]
    print("entropy = ", p1)
    s2 = curr_feat_n + "_gain_r"
    p2 = globals()[s2]
    print("gain ratio = ", p2)
    print("feature = ", curr_feat_n)
    for it in dit.keys() :
        decision(dit[it], idx+1)
```
