```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("0000000000002429_training_titanic_x_y_train.csv")
val = data.Age.mean()
data.Age = data.Age.fillna(val)
x_train = data.iloc[:, [3, 4, 5]].values
y_train = data.iloc[:, 10].values
x_train
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(x_train, y_train)
```

```python
testdata = pd.read_csv("0000000000002429_test_titanic_x_test.csv")
val2 = testdata.Age.mean()
testdata.Age = testdata.Age.fillna(val2)
x_test = testdata.iloc[:, [3, 4, 5]].values
x_test
```

```python
y_pred = classifier.predict(x_test)
y_pred
```
