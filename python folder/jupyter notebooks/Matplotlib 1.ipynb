```python
import matplotlib.pyplot as plt
```

```python
x=[1, 2, 3]
y=[2, 4, 6]

# y=2x line
plt.scatter(x, y)
plt.show()
```

```python
x=[1, 2, 3]
y=[2, 4, 6]

# y=2x line
plt.plot(x, y)
plt.show()
```

```python
x=[1, 2, 3, 4, 6]
y=[2, 4, 6, 8, 12]

x1=[1, 4, 5, 6]
y1=[3, 7, 10, 11]

plt.scatter(x1, y1)
plt.plot(x1, y1)

plt.scatter(x, y)
plt.plot(x, y)

plt.show()
```

```python
x1=[1, 4, 5, 6]
y1=[3, 7, 10, 11]

# changing the colour and style, in the plot function

plt.scatter(x1, y1)
plt.plot(x1, y1, "g-.")
plt.show()
```

```python
# plot function x^3

import numpy as np

# x=np.array([1, 2, 3, 4])
x=np.arange(0, 16, 0.01) # (starting point, ending point, step)
y=x**3
plt.plot(x, y)
plt.show()
```

```python
x=np.arange(0, 16, 0.01) # (starting point, ending point, step)

y=x**3
y1=x**2

plt.plot(x, y, color="green", label="cube")
plt.plot(x, y1, color="red", label="square")

plt.ylabel("Y")
plt.xlabel("X")

plt.legend() # to show the labels

plt.title("matplotlib demo")

plt.show()
```

```python
x=np.arange(0, 16, 0.01) # (starting point, ending point, step)

y=x**3
y1=x**2

plt.plot(x, y, color="green", label="cube")
plt.plot(x, y1, color="red", label="square")

plt.ylabel("Y")
plt.xlabel("X")

plt.legend() # to show the labels

plt.title("matplotlib demo")

plt.grid() # creates a grid, super helpful

plt.show()
```

```python
# printing text at a given point

x=np.arange(0, 16, 0.01) # (starting point, ending point, step)

y=x**3
y1=x**2

plt.plot(x, y, color="green", label="cube")
plt.plot(x, y1, color="red", label="square")

plt.ylabel("Y")
plt.xlabel("X")

plt.legend() # to show the labels
plt.title("matplotlib demo")

plt.text(2, 1000, "arya")

plt.grid() # creates a grid, super helpful

plt.show()
```
