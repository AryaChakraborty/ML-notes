## Pie Graphs

```python
import matplotlib.pyplot as plt
```

```python
sizes = [3, 4, 6, 11]
plt.pie(sizes)
plt.show()
```

```python
# if the size of the pie graph is not ok, then we can do...
# plt.axis("equal")
```

```python
sizes = [3, 4, 6, 11]
label = ["A", "B", "C", "D"]
colour=["pink", "purple", "red", "orange"]
plt.pie(sizes, colors=colour, labels=label)
plt.show()
```

```python
# autopct function

sizes = [3, 4, 6, 11]
label = ["A", "B", "C", "D"]
colour=["pink", "purple", "red", "orange"]
plt.pie(sizes, colors=colour, labels=label, autopct="%.2f%%") # %% is to show thw % sign, and 2 is 2 decimal places
plt.show()
```

```python
# highlighting a slice of a pie

sizes = [3, 4, 6, 11]
label = ["A", "B", "C", "D"]
explode=[0.5, 0, 0.1, 0]
colour=["pink", "purple", "red", "orange"]
plt.pie(sizes, colors=colour, labels=label, autopct="%.2f%%", explode=explode)
plt.show()
```
