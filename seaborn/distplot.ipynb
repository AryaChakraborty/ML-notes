```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = sns.load_dataset('titanic')
data
```

```python
corr_tab = data.corr()
corr_tab
```

```python
sns.heatmap(corr_tab, cmap='Reds')
```

```python
# probability dist of age count

sns.distplot(data['age'])
```

```python
sns.distplot(data['age'], hist=False)
```

```python
sns.distplot(data['age'], kde=False)
```

```python
sns.distplot(data['age'], kde=False, hist=False, rug=True)
```

```python
# age distribution of people who survived

sns.distplot(data[data['survived'] == 1]['age'])
```

```python
# age distribution of people who survived vs died

sns.distplot(data[data['survived'] == 1]['age'])
sns.distplot(data[data['survived'] == 0]['age'])
```
