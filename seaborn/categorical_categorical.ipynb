```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = sns.load_dataset('flights')
data
```

```python
corr_mat = data.corr()
corr_mat
```

```python
sns.heatmap(corr_mat, cmap='Reds')
```

## let's check how many passengers flew in a month for a year

```python
table = data.pivot_table(index='year', columns='month', values='passengers', aggfunc='sum')
table
```

```python
sns.heatmap(table, cmap='Reds')
```

```python
sns.heatmap(table, annot=True, fmt='d', linewidth=0.5)
```

# clustermap

```python
sns.clustermap(table)
```
