```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = pd.read_csv('tips.csv')
data
```

# bar plot

```python
sns.catplot(x='smoker', y='total_bill', kind='bar', data=data)
```

```python
sns.catplot(x='smoker', y='total_bill', kind='bar', hue='sex', data=data)
```

```python
sns.catplot(x='smoker', y='total_bill', kind='bar', hue='sex', estimator=np.median, data=data)
```

```python
sns.catplot(x='smoker', y='total_bill', kind='bar', hue='sex', estimator=np.var, data=data)
```

```python
sns.catplot(x='smoker', y='total_bill', kind='bar', hue='sex', estimator=np.std, data=data)
```

# countplots

```python
sns.catplot(x='sex', kind='count', data=data)
```

```python
sns.catplot(x='sex', hue='smoker', kind='count', data=data)
```
